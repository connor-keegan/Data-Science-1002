```python
# import necessary packages
import numpy as np
from PIL import Image
import requests
import matplotlib.pyplot as plt

# get image from user
url = input("JPG Image url: ")
image = np.asarray(Image.open(requests.get(url, stream=True).raw))

# function to make the image red
def make_photo_red(image):
  red_photo_first = np.copy(image)
  red_photo_first[:,:,1] = 0
  red_photo_second = np.copy(red_photo_first)
  red_photo_second[:,:,2] = 0
  return red_photo_second

# function to make the image green
def make_photo_green(image):
  green_photo_first = np.copy(image)
  green_photo_first[:,:,0] = 0
  green_photo_second = np.copy(green_photo_first)
  green_photo_second[:,:,2] = 0
  return green_photo_second

# function to make the image blue
def make_photo_blue(image):
  blue_photo_first = np.copy(image)
  blue_photo_first[:,:,0] = 0
  blue_photo_second = np.copy(blue_photo_first)
  blue_photo_second[:,:,1] = 0
  return blue_photo_second

# store all colored images in variables
red_image = make_photo_red(image)
green_image = make_photo_green(image)
blue_image = make_photo_blue(image)

# make a version of the image that is red and green without blue
red_and_green_image = np.copy(red_image)
green_image_copy = np.copy(green_image)
red_and_green_image[:,:,1] = green_image_copy[:,:,1]
red_and_green_final = np.copy(red_and_green_image)

# make a version of the image that is red and blue with no green
red_and_blue_image = np.copy(red_image)
blue_image_copy = np.copy(blue_image)
red_and_blue_image[:,:,2] = blue_image_copy[:,:,2]
red_and_blue_final = np.copy(red_and_blue_image)

# make a version of the image that is green and blue with no red
green_and_blue_image = np.copy(green_image)
blue_image_copy_two = np.copy(blue_image)
green_and_blue_image[:,:,2] = blue_image_copy_two[:,:,2]
green_and_blue_final = np.copy(green_and_blue_image)

# combine the red, green, and blue images into a complete photo
complete_photo = np.copy(red_image) + np.copy(green_image) + np.copy(blue_image)

# print and label the original image
print("")
print("Your original image: ")
plt.imshow(image)
plt.show()
plt.figure()

# print and label the image in red
print("")
print("Your image in red: ")
plt.imshow(red_image)
plt.show()
plt.figure()

# print and label the image in green
print("")
print("Your image in green: ")
plt.imshow(green_image)
plt.show()
plt.figure()

# print and label image in blue
print("")
print("Your image in blue: ")
plt.imshow(blue_image)
plt.show()
plt.figure()

# print and label image in red and green
print("")
print("Your image in red and green: ")
plt.imshow(red_and_green_final)
plt.show()
plt.figure()

#print image in red and blue
print("")
print("Your image in red and blue: ")
plt.imshow(red_and_blue_final)
plt.show()
plt.figure()

#print image in green and blue
print("")
print("Your image in green and blue: ")
plt.imshow(green_and_blue_final)
plt.show()
plt.figure()

#check to see if complete photo is the same as the original photo
print("")
if np.all(complete_photo) == np.all(image):
  print("Success! The original image is identical to the image obtained by combining the red, green, and blue images.")
else:
  print("Failure! The originial image is not identical to the image obtained by combining the red, green, and blue images.")
```
